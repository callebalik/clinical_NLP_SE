```python
# Matcher
from spacy.matcher import PhraseMatcher
from spacy import displacy
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
# Only run nlp.make_doc to speed things up
covid19patterns = [nlp.make_doc(text) for text in sample_phrases]
#create matcher object,the first argument is the class
matcher.add("covid19", None, *covid19patterns)
```

```python
# Doc
doc = nlp(abstract)
matches = matcher(doc)
entities = []

for match_id, start, end in matches:
    span = doc[start:end]
    class_id = nlp.vocab.strings[match_id]
    print(span.text, span.start_char, span.end_char, class_id)
    ent = {"start": span.start_char, "end": span.end_char, "label": class_id}
    entities.append(ent)
print(entities)
```
